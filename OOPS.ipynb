How is encapsulation achieved in Python


Encapsulation: This is about bundling data (attributes) and methods (functions) that operate on the data into a single unit (a class). It also involves restricting direct access to some of an object's components (data hiding), meaning that the internal state of an object is protected from direct external access. You interact with the object through its public methods. Think of a car's engine: you don't directly manipulate the pistons; you use the accelerator and brake pedals (public methods) to control it.

In Python, encapsulation is achieved through conventions rather than strict access modifiers. Here's how it generally works:

### 1. Public Members

````python
class MyClass:
    def __init__(self, public_attribute):
        self.public_attribute = public_attribute

    def public_method(self):
        return f"This is a public method. Public attribute: {self.public_attribute}"

obj = MyClass("I am public")
print(obj.public_attribute)
print(obj.public_method())
````

### 2. Protected Members (Convention)

By convention, a single leading underscore `_` indicates that an attribute or method is intended for internal use within the class or its subclasses. It's a hint to other developers not to access it directly, but Python does not strictly enforce this.

````python
class MyClass:
    def __init__(self, protected_attribute):
        self._protected_attribute = protected_attribute

    def _protected_method(self):
        return f"This is a protected method. Protected attribute: {self._protected_attribute}"

obj = MyClass("I am protected")
print(obj._protected_attribute) # Accessible, but discouraged
print(obj._protected_method())  # Accessible, but discouraged
````

### 3. Private Members (Name Mangling)

A double leading underscore `__` triggers a mechanism called **name mangling**. Python renames the attribute internally to `_ClassName__attributeName`. This makes it harder (but not impossible) to access from outside the class, providing a stronger form of encapsulation.

````python
class MyClass:
    def __init__(self, private_attribute):
        self.__private_attribute = private_attribute

    def get_private_attribute(self):
        return self.__private_attribute

obj = MyClass("I am private")
# print(obj.__private_attribute) # This would raise an AttributeError
print(obj.get_private_attribute()) # Access via public method (encapsulated)

# Accessing via name mangling (discouraged, but possible)
print(obj._MyClass__private_attribute)
````

In summary, while Python doesn't have explicit `private` keywords, it relies on conventions and name mangling to achieve encapsulation, encouraging developers to interact with objects through their defined public interfaces.

What is a constructor in Python


In Python, a **constructor** is a special method that gets automatically called when you create a new object (an instance) of a class. Its primary purpose is to initialize the attributes of the new object.

The constructor method in Python is always named `__init__` (with double underscores before and after). It takes `self` as its first parameter, which refers to the newly created instance of the class. Any additional parameters are used to provide initial values for the object's attributes.

````python
class Dog:
    # This is the constructor method
    def __init__(self, name, breed):
        self.name = name  # Initialize the 'name' attribute
        self.breed = breed # Initialize the 'breed' attribute
        print(f"A new dog named {self.name} of breed {self.breed} has been created!")

    def bark(self):
        return f"{self.name} says Woof!"

# Creating objects (instances) of the Dog class
# The __init__ method is automatically called here
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Lucy", "Labrador")

print(dog1.name)
print(dog2.breed)
print(dog1.bark())
````

In this example:
- `__init__(self, name, breed)` is the constructor.
- When `dog1 = Dog("Buddy", "Golden Retriever")` is called, the `__init__` method is executed, setting `dog1.name` to "Buddy" and `dog1.breed` to "Golden Retriever".

What are class and static methods in Python


## Class Methods

A **class method** is a method that is bound to the class and not the instance of the class. It receives the class itself as its first argument (conventionally named `cls`), instead of the instance (`self`). Class methods are defined using the `@classmethod` decorator.

### When to use class methods:
- **Factory Methods:** To create instances of the class in different ways.
- **Alternative Constructors:** When you need multiple ways to initialize an object.
- **Modifying Class State:** To perform operations that modify a class variable that applies to all instances.

````python
class Car:
    wheels = 4  # Class variable

    def __init__(self, make, model):
        self.make = make
        self.model = model

    def display_info(self):
        return f"This is a {self.make} {self.model} with {self.wheels} wheels."

    @classmethod
    def with_custom_wheels(cls, make, model, num_wheels):
        car = cls(make, model) # Create an instance of the class
        car.wheels = num_wheels # Modify instance-specific 'wheels' (or class if not instance)
        return car

    @classmethod
    def change_default_wheels(cls, new_num_wheels):
        cls.wheels = new_num_wheels # Modifies the class variable 'wheels'
        return f"Default car now has {cls.wheels} wheels."

# Using instance method
car1 = Car("Toyota", "Camry")
print(car1.display_info())

# Using class method as a factory
car2 = Car.with_custom_wheels("Honda", "Civic", 3)
print(car2.display_info()) # This will show 3 wheels for car2

# What about car1? It still has 4 wheels as 'wheels' was set on the instance
print(car1.display_info())

# Change the class-level default wheels
print(Car.change_default_wheels(6))

# Now, new cars will have 6 wheels by default
car3 = Car("Ford", "Focus")
print(car3.display_info())

# car1 and car2 are unaffected by the class-level change unless they didn't have their own 'wheels' attribute
# Since with_custom_wheels explicitly sets car.wheels, car2 is still 3
# car1 also has its own 'wheels' attribute set during its initialization or earlier, if not, it would inherit
````

## Static Methods

A **static method** is a method that belongs to a class but doesn't operate on the instance or the class itself. It doesn't receive `self` or `cls` as its first argument. Static methods are defined using the `@staticmethod` decorator.

### When to use static methods:
- **Utility Functions:** When a method has a logical connection to the class but doesn't need to access any instance or class-specific data.
- **Helper Functions:** For functions that perform independent tasks but are conceptually related to the class.

````python
import datetime

class MathOperations:
    def __init__(self, value):
        self.value = value

    @staticmethod
    def add_numbers(x, y):
        return x + y

    @staticmethod
    def is_weekend(date):
        return date.weekday() >= 5  # Saturday (5) or Sunday (6)

    def multiply_by_self(self):
        return self.value * self.value

# Using static methods directly through the class or an instance
print(f"Sum: {MathOperations.add_numbers(5, 7)}")

today = datetime.date(2023, 10, 28) # A Saturday
print(f"Is {today} a weekend? {MathOperations.is_weekend(today)}")

# Using an instance method
calc = MathOperations(10)
print(f"Multiply by self: {calc.multiply_by_self()}")
````

## Key Differences:

| Feature             | Instance Methods                 | Class Methods                    | Static Methods                   |
|---------------------|----------------------------------|----------------------------------|----------------------------------|
| **First Argument**  | `self` (instance)                | `cls` (class)                    | None (`self` or `cls`)           |
| **Decorator**       | None (default)                   | `@classmethod`                   | `@staticmethod`                  |
| **Access to State** | Can access/modify instance state | Can access/modify class state    | Cannot access instance or class state |
| **Use Case**        | Object-specific operations       | Factory methods, class-level logic | Utility functions, no state dependency |
| **Calling**         | `obj.method()`                   | `Class.method()` or `obj.method()` | `Class.method()` or `obj.method()` |


What is method overloading in Python


Method overloading, in the context of Object-Oriented Programming, refers to the ability to define multiple methods within the same class that have the same name but different parameters (e.g., different number of arguments or different types of arguments). The appropriate method is then called based on the arguments passed during the method call.

However, Python does not support method overloading in the traditional sense as seen in languages like Java or C++. If you define multiple methods with the same name in a Python class, the later definition will simply override the earlier ones. Only the last defined method with that name will be available.



What is method overriding in OOP


## What is Method Overriding in OOP?

Method overriding is a feature in Object-Oriented Programming (OOP) that allows a subclass or child class to provide a specific implementation of a method that is already provided by one of its superclasses or parent classes.

When a method in a subclass has the same name, same parameters (or signature), and same return type (or compatible return type) as a method in its superclass, then the method in the subclass is said to **override** the method in the superclass.

### Key Principles:
1.  **Inheritance:** Overriding relies on an inheritance relationship, where a subclass inherits from a superclass.
2.  **Same Signature:** The overriding method in the subclass must have the same name and the same number and types of parameters as the method it overrides in the superclass.
3.  **Specific Implementation:** The purpose of overriding is to provide a specialized or different behavior for a method in the subclass, while still maintaining the common interface defined by the superclass.
4.  **Polymorphism:** Method overriding is a crucial aspect of runtime polymorphism (dynamic polymorphism). When an object of a subclass is referred to by a superclass reference, and an overridden method is called, the version of the method in the actual object's class (the subclass) is executed.

### Example in Python:

````python
class Animal:
    def speak(self):
        return "The animal makes a sound."

class Dog(Animal):
    # This method overrides the speak method from the Animal class
    def speak(self):
        return "Woof! Woof!"

class Cat(Animal):
    # This method also overrides the speak method
    def speak(self):
        return "Meow!"

# Creating instances
animal = Animal()
dog = Dog()
cat = Cat()

# Calling the speak method on different objects
print(f"Animal says: {animal.speak()}")
print(f"Dog says: {dog.speak()}")
print(f"Cat says: {cat.speak()}")

# Demonstrating polymorphism
animals = [Animal(), Dog(), Cat()]
for a in animals:
    print(f"Polymorphic call: {a.speak()}")
````

In this example:
- The `Animal` class has a `speak` method.
- The `Dog` and `Cat` classes inherit from `Animal`.
- Both `Dog` and `Cat` classes provide their own `speak` method, which has the same name and no parameters, thus **overriding** the `speak` method inherited from `Animal`.
- When we call `speak()` on a `Dog` object, the `Dog`'s version is executed. Similarly for the `Cat` object.
- The polymorphic example shows how different objects respond to the same method call (`a.speak()`) based on their actual type, demonstrating the power of method overriding.

What is a property decorator in Python


A **property decorator** (`@property`) in Python is a built-in decorator that provides a Pythonic way to access attributes of an object. It allows you to define methods within a class that can be accessed like regular attributes, rather than having to call them as functions. This is primarily used for:

1.  **Encapsulation and Data Validation:** It helps in encapsulating internal implementation details and provides a controlled way to get, set, or delete an attribute's value. You can add logic for validation or transformation when an attribute is accessed or modified.
2.  **Read-only Attributes:** You can create attributes that are read-only by only defining the getter method.
3.  **Backward Compatibility:** If you initially expose a public attribute but later decide that some logic needs to be executed upon its access or modification, you can convert it into a property without changing the way external code interacts with it.

The `@property` decorator is typically used in conjunction with three methods:

*   **`@property` (getter):** Marks a method as a getter for an attribute.
*   **`@attribute.setter`:** Marks a method as a setter for the attribute named `attribute`.
*   **`@attribute.deleter`:** Marks a method as a deleter for the attribute named `attribute`.

### Example:

````python
class Person:
    def __init__(self, name, age):
        self._name = name  # Private convention for internal storage
        self._age = None   # Initialize private attribute
        self.age = age     # Use the setter to set the age initially

    @property
    def name(self):
        """The name property (getter)."""
        print("Getting name...")
        return self._name

    @name.setter
    def name(self, value):
        """The name setter."""
        print("Setting name...")
        if not isinstance(value, str) or not value.strip():
            raise ValueError("Name must be a non-empty string")
        self._name = value.strip()

    @property
    def age(self):
        """The age property (getter)."""
        print("Getting age...")
        return self._age

    @age.setter
    def age(self, value):
        """The age setter with validation."""
        print("Setting age...")
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError("Age must be a non-negative number")
        self._age = int(value) # Ensure age is an integer

    @age.deleter
    def age(self):
        """The age deleter."""
        print("Deleting age...")
        del self._age

# Create an instance
p = Person("Alice", 30)

# Accessing attributes (calls the getter)
print(f"Person's name: {p.name}")
print(f"Person's age: {p.age}")

# Setting attributes (calls the setter with validation)
p.name = "Bob Smith"
p.age = 25
print(f"Updated name: {p.name}")
print(f"Updated age: {p.age}")

# Attempt to set invalid age (will raise ValueError)
try:
    p.age = -5
except ValueError as e:
    print(f"Error setting age: {e}")

# Deleting an attribute (calls the deleter)
del p.age

# Attempt to access deleted attribute (will raise AttributeError)
try:
    print(p.age)
except AttributeError as e:
    print(f"Error accessing deleted age: {e}")
````

In this example:
- `name` and `age` are exposed as properties.
- When you access `p.name` or `p.age`, the respective `@property` getter method is called.
- When you assign a value like `p.name = "Bob"` or `p.age = 25`, the `@name.setter` or `@age.setter` method is called, allowing for custom logic (like validation) before the internal `_name` or `_age` attribute is updated.
- When you use `del p.age`, the `@age.deleter` method is called to handle the deletion logic.

Why is polymorphism important in OOP


Polymorphism, which means "many forms," is a fundamental concept in Object-Oriented Programming (OOP) that allows objects of different classes to be treated as objects of a common type. It enables you to define one interface (e.g., a method name) and have multiple implementations of it, depending on the specific class of the object.

### Why is Polymorphism Important in OOP?

1.  **Code Reusability:** You can write generic code that works with objects of different classes, as long as those classes share a common interface (e.g., inherit from a common base class or implement a common interface/protocol). This reduces redundancy and makes your code more DRY (Don't Repeat Yourself).

2.  **Flexibility and Extensibility:** Polymorphism makes it easy to add new classes to a system without modifying existing code. If new classes adhere to the common interface, they can be seamlessly integrated and used by existing polymorphic code. This is crucial for systems that need to evolve and adapt.

3.  **Maintainability:** Changes or bug fixes in a specific implementation can be isolated to that class, without affecting the generic code that uses the polymorphic interface. This simplifies maintenance and reduces the risk of introducing new bugs.

4.  **Readability and Simplicity:** By interacting with objects through a common interface, the code becomes more abstract and easier to understand. You focus on 'what' an object does (its behavior) rather than 'how' it does it (its specific implementation details).

5.  **Decoupling:** Polymorphism helps in decoupling the client code (the code that uses the objects) from the specific implementations of the objects. This makes the system more modular and less prone to cascading changes.

6.  **Dynamic Binding (Runtime Polymorphism):** The decision of which specific method implementation to call is made at runtime, based on the actual type of the object. This allows for dynamic and flexible behavior.

### Example (revisiting the `Animal` example):

Consider our `Animal`, `Dog`, and `Cat` classes with their `speak` method.

```python
class Animal:
    def speak(self):
        return "The animal makes a sound."

class Dog(Animal):
    def speak(self):
        return "Woof! Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

def make_animal_speak(animal):
    # This function doesn't need to know the specific type of animal
    # It just knows that 'animal' has a 'speak' method
    print(animal.speak())

my_dog = Dog()
my_cat = Cat()
my_animal = Animal()

make_animal_speak(my_dog)   # Output: Woof! Woof!
make_animal_speak(my_cat)   # Output: Meow!
make_animal_speak(my_animal) # Output: The animal makes a sound.

# If we later add a 'Cow' class:
class Cow(Animal):
    def speak(self):
        return "Moo!"

my_cow = Cow()
make_animal_speak(my_cow) # Output: Moo!
```

In this example, the `make_animal_speak` function can handle any object that inherits from `Animal` and has a `speak` method, without needing to be modified each time a new type of `Animal` is introduced. This demonstrates the power of polymorphism in creating flexible and extensible code.

What is an abstract class in Python


## What is an Abstract Class in Python?

An **abstract class** is a class that cannot be instantiated on its own and is designed to be subclassed. It acts as a blueprint or a contract for other classes, defining a common interface that its concrete subclasses must implement. Abstract classes are a core concept in Object-Oriented Programming (OOP) for achieving abstraction.

### Key Characteristics:
1.  **Cannot be Instantiated:** You cannot create an object directly from an abstract class. You must create a subclass that implements its abstract methods.
2.  **Abstract Methods:** Abstract classes can declare one or more abstract methods. An **abstract method** is a method that has a declaration but no implementation. Subclasses are *required* to provide an implementation for all inherited abstract methods.
3.  **Enforces Interface:** They enforce a consistent interface among a group of related classes. Any class that inherits from an abstract class must provide implementations for its abstract methods, ensuring a certain level of functionality.
4.  **Promotes Polymorphism:** Abstract classes contribute to polymorphism by allowing a base class reference to refer to objects of its concrete subclasses, and when an abstract method is called, the subclass's implementation is executed.

### How Python Implements Abstract Classes:
Python does not have abstract classes built-in like some other languages (e.g., Java's `abstract` keyword). Instead, it provides the `abc` module (Abstract Base Classes) to define abstract classes and abstract methods.

-   To make a class an abstract base class, you inherit from `ABC` from the `abc` module.
-   To define an abstract method, you use the `@abstractmethod` decorator from the `abc` module.

### Example:

````python
from abc import ABC, abstractmethod

class Shape(ABC): # Inherit from ABC to make it an abstract base class
    @abstractmethod
    def area(self):
        """Abstract method to calculate the area. Must be implemented by subclasses."""
        pass # No implementation here

    @abstractmethod
    def perimeter(self):
        """Abstract method to calculate the perimeter. Must be implemented by subclasses."""
        pass # No implementation here

    def describe(self):
        """A concrete method that can be implemented in the abstract class or overridden in subclasses."""
        return "This is a generic shape."

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14159 * self.radius

    def describe(self):
        return f"This is a circle with radius {self.radius}."

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

# --- Usage Examples ---

# Attempting to instantiate an abstract class will raise an error
try:
    # generic_shape = Shape() # This line would raise a TypeError
    print("Cannot instantiate abstract class Shape directly.")
except TypeError as e:
    print(f"Error: {e}")

# Instantiate concrete subclasses
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"\nCircle description: {circle.describe()}")
print(f"Circle area: {circle.area()}")
print(f"Circle perimeter: {circle.perimeter()}")

print(f"\nRectangle description: {rectangle.describe()}")
print(f"Rectangle area: {rectangle.area()}")
print(f"Rectangle perimeter: {rectangle.perimeter()}")

# Polymorphism: treating different shapes as a common 'Shape' type
shapes = [circle, rectangle]
for s in shapes:
    print(f"\n--- {s.describe()} ---")
    print(f"Area: {s.area()}")
    print(f"Perimeter: {s.perimeter()}")
````

In this example:
- `Shape` is an abstract base class because it inherits from `ABC` and defines abstract methods `area()` and `perimeter()` using `@abstractmethod`.
- You cannot create an object directly from `Shape`.
- `Circle` and `Rectangle` are concrete subclasses that inherit from `Shape`. They *must* provide implementations for `area()` and `perimeter()`, otherwise, they themselves would become abstract classes (and attempting to instantiate them would raise an error).
- The `describe()` method is a concrete method in the abstract class, which can be used as-is or overridden by subclasses.
- The example demonstrates how polymorphism works: both `Circle` and `Rectangle` objects can be treated as `Shape` objects, and their specific `area()` and `perimeter()` implementations are called dynamically.

What is the difference between a class variable and an instance variable


## Class Variables vs. Instance Variables in Python

In Object-Oriented Programming (OOP) in Python, both class variables and instance variables are used to store data within a class. However, they differ significantly in their scope, ownership, and how their values are accessed and modified.

### 1. Instance Variables

-   **Definition:** Instance variables are owned by instances (objects) of a class. Each object has its own copy of the instance variables.
-   **Declaration:** They are typically defined inside the constructor method (`__init__`) using the `self` keyword.
-   **Scope:** Specific to a particular object. Changes to an instance variable in one object do not affect the same variable in other objects.
-   **Use Case:** To store data that is unique to each instance of the class.

### 2. Class Variables

-   **Definition:** Class variables are owned by the class itself, not by any specific instance. There is only one copy of a class variable, and it is shared among all instances of the class.
-   **Declaration:** They are typically defined directly inside the class but outside of any method.
-   **Scope:** Accessible by all instances of the class, and also directly through the class name.
-   **Use Case:** To store data that is common to all instances of the class, or to define constants that belong to the class.

### Key Differences:

| Feature           | Instance Variable                          | Class Variable                                  |
| :---------------- | :----------------------------------------- | :---------------------------------------------- |
| **Ownership**     | Belongs to each instance (object)          | Belongs to the class itself                     |
| **Copies**        | Each instance has its own copy             | One copy, shared by all instances               |
| **Definition**    | Inside `__init__` with `self.name`         | Directly inside the class, outside methods      |
| **Access**        | `instance.name`                            | `ClassName.name` or `instance.name`             |
| **Modification**  | Modifying `instance.name` affects only that instance | Modifying `ClassName.name` affects all instances; modifying `instance.name` (if no instance variable of that name exists yet) creates an instance variable that shadows the class variable for that instance. |
| **Purpose**       | Unique data for each object                | Shared data/constants for all objects           |


````python
class Car:
    # Class variable: shared among all instances
    wheels = 4
    # Another class variable
    vehicle_type = "Automobile"

    def __init__(self, make, model, color):
        # Instance variables: unique to each instance
        self.make = make
        self.model = model
        self.color = color

    def display_info(self):
        return f"A {self.color} {self.make} {self.model} with {self.wheels} wheels ({self.vehicle_type})."

# Create instances of the Car class
car1 = Car("Toyota", "Camry", "Blue")
car2 = Car("Honda", "Civic", "Red")

print("--- Initial State ---")
print(f"Car 1: {car1.display_info()}")
print(f"Car 2: {car2.display_info()}")
print(f"Class variable 'wheels' (via class): {Car.wheels}")
print(f"Class variable 'wheels' (via instance 1): {car1.wheels}")
print(f"Class variable 'wheels' (via instance 2): {car2.wheels}")

# Modify an instance variable
car1.color = "Green"
print("\n--- After modifying instance variable 'color' for car1 ---")
print(f"Car 1: {car1.display_info()}")
print(f"Car 2: {car2.display_info()}") # car2's color remains unaffected

# Modify a class variable directly via the class
Car.wheels = 6
print("\n--- After modifying class variable 'wheels' via Class ---")
print(f"Car 1: {car1.display_info()}") # Both instances now reflect the change
print(f"Car 2: {car2.display_info()}")
print(f"Class variable 'wheels' (via class): {Car.wheels}")

# What happens if we try to modify a class variable via an instance?
# Python will create a new instance variable that shadows the class variable for that instance.
car1.wheels = 8 # This creates a new instance variable 'wheels' for car1
print("\n--- After modifying 'wheels' via car1 instance ---")
print(f"Car 1: {car1.display_info()}") # Car1 now uses its own 'wheels' (8)
print(f"Car 2: {car2.display_info()}") # Car2 still uses the class 'wheels' (6)
print(f"Class variable 'wheels' (via class): {Car.wheels}") # The class variable is still 6
print(f"Car1's own wheels: {car1.wheels}")
print(f"Car2's own wheels: {car2.wheels}")
````

What is multiple inheritance in Python


## What is Multiple Inheritance in Python?

**Multiple inheritance** is a feature in Object-Oriented Programming (OOP) where a class can inherit properties and behaviors (attributes and methods) from more than one parent (or base) class. This allows a new class to combine functionalities from multiple distinct sources.

### How it works in Python:

In Python, a class can inherit from several parent classes by listing them in the class definition's parentheses, separated by commas:

```python
class ChildClass(ParentClass1, ParentClass2, ParentClass3):
    # ... class body ...
```

### The Diamond Problem and Method Resolution Order (MRO):

One of the main challenges with multiple inheritance is the "Diamond Problem" (or "Deadly Diamond of Death"). This occurs when a class inherits from two classes that have a common ancestor, and all three classes implement the same method. The question then arises: which parent's method should be inherited by the child class?

Python handles this complexity using a well-defined algorithm called the **Method Resolution Order (MRO)**. The MRO determines the order in which base classes are searched when a method is called on an object. Python 2 and Python 3 use different MRO algorithms:

-   **Python 2 (Classic Classes):** Used a depth-first, left-to-right algorithm, which could lead to unpredictable results in complex hierarchies.
-   **Python 3 (New-style Classes):** Uses the C3 Linearization algorithm. This algorithm ensures that the MRO is consistent, monotonic (if a method is found in a superclass, it will always be found there or further down the hierarchy), and satisfies local precedence (a class is always searched before its parents, and listed parent classes are searched from left to right).

You can inspect the MRO of any class using its `__mro__` attribute or the `help()` function.

### Advantages of Multiple Inheritance:

-   **Code Reusability:** Allows a class to reuse code from multiple parent classes.
-   **Increased Functionality:** A child class can gain a wider range of functionalities by combining behaviors from various sources.
-   **Flexibility:** Provides a way to design more flexible and modular systems.

### Disadvantages and Considerations:

-   **Complexity:** Can make class hierarchies difficult to understand and manage.
-   **Ambiguity:** Increases the chances of method name clashes (though MRO resolves which one is called, it can still lead to confusion).
-   **Design Challenges:** Often leads to tighter coupling between classes if not designed carefully.

Due to these potential complexities, many developers prefer using **composition over inheritance** or limiting multiple inheritance to mixin classes (classes designed to provide specific functionalities to other classes, not meant to be standalone).


### Example of Multiple Inheritance and MRO:

Let's consider an example where a `FlyingCar` inherits from both `Car` and `Aircraft` classes.

````python
class Car:
    def __init__(self, brand):
        self.brand = brand

    def drive(self):
        return f"{self.brand} car is driving."

    def honk(self):
        return f"{self.brand} car honks: Beep! Beep!"

class Aircraft:
    def __init__(self, model):
        self.model = model

    def fly(self):
        return f"{self.model} aircraft is flying."

    def honk(self):
        return f"{self.model} aircraft sounds: WHOOSH!"

class FlyingCar(Car, Aircraft):
    def __init__(self, brand, model):
        Car.__init__(self, brand) # Initialize Car part
        Aircraft.__init__(self, model) # Initialize Aircraft part

    def take_off(self):
        return f"The {self.brand} {self.model} is taking off!"

    # The 'honk' method from Car is prioritized due to MRO
    # If you wanted Aircraft's honk, you'd explicitly call it or put Aircraft first

# Create an instance of FlyingCar
my_flying_car = FlyingCar("Tesla", "SkyPod")

print(my_flying_car.drive()) # Inherited from Car
print(my_flying_car.fly())   # Inherited from Aircraft
print(my_flying_car.take_off()) # Defined in FlyingCar

# Demonstrating Method Resolution Order (MRO)
# Which honk method will be called?
print(my_flying_car.honk()) # Calls Car's honk because Car is listed first in the inheritance

# Inspect the MRO
print("\nMRO for FlyingCar:")
print(FlyingCar.__mro__)

# Another way to view MRO
print("\nHelp for FlyingCar:")
help(FlyingCar)
````

Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python


## The purpose of `__str__` and `__repr__` methods in Python

In Python, `__str__` and `__repr__` are special methods (dunder methods) used to define string representations of objects. While both return strings, they are intended for different audiences and serve different purposes.

### 1. `__repr__` (Representation)

-   **Purpose:** The primary goal of `__repr__` is to provide an **unambiguous** string representation of an object. It's mainly for developers for debugging and logging.
-   **Audience:** Developers.
-   **Convention:** The string returned by `__repr__` should ideally be a valid Python expression that could be used to recreate the object (if feasible). For example, `eval(repr(obj))` should ideally return a new object equal to `obj`.
-   **When it's called:**
    -   By the `repr()` built-in function.
    -   By the interactive interpreter when an object is displayed as the output of an expression (if `__str__` is not defined).
    -   For debugging in IDEs.

### 2. `__str__` (String)

-   **Purpose:** The primary goal of `__str__` is to provide a **readable** and **user-friendly** string representation of an object. It's meant for the end-user.
-   **Audience:** End-users, for displaying information.
-   **Convention:** The string returned by `__str__` should be concise, informative, and easy to understand, even if it doesn't contain all the details needed to recreate the object.
-   **When it's called:**
    -   By the `str()` built-in function.
    -   By the `print()` function.
    -   By `f-strings` or the `format()` method.
    -   Implicitly when converting an object to a string for display in various contexts (e.g., writing to a file, GUI elements).

### Relationship and Fallback:

-   If a class defines `__repr__` but not `__str__`, then `__repr__` will be used as the string representation when `str()` or `print()` is called.
-   If a class defines `__str__` but not `__repr__`, then only `__str__` will be used by `str()` and `print()`. When `repr()` is called or the object is displayed in the interpreter, it will fall back to the default `__repr__` provided by Python (e.g., `<MyClass object at 0x...>`).
-   It's generally a good practice to define at least `__repr__` for all your custom classes, as it's crucial for debugging. Defining `__str__` is optional but recommended if you need a user-friendly string representation different from the developer-oriented one.


### Example:

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        # Developer-friendly, unambiguous representation
        # Ideally, `eval(repr(point))` should recreate the object
        return f"Point(x={self.x}, y={self.y})"

    def __str__(self):
        # User-friendly, readable representation
        return f"({self.x}, {self.y})"

# Create an instance of the Point class
p = Point(10, 20)

print("--- Using an instance with both __str__ and __repr__ ---")
# print() calls __str__
print(p)

# str() calls __str__
print(str(p))

# repr() calls __repr__
print(repr(p))

# Interactive interpreter output (calls __repr__ if __str__ is not present,
# otherwise it calls __repr__ if it's the last expression)
# In a script, this just prints the repr if it's the last line, or nothing otherwise.
print(f"Direct object in output (would show __repr__ in interpreter): {p!r}")

print("\n--- Example with only __repr__ defined ---")
class AnotherPoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"AnotherPoint(x={self.x}, y={self.y})"

ap = AnotherPoint(5, 7)
print(ap)       # Calls __repr__ because __str__ is not defined
print(str(ap))  # Calls __repr__ because __str__ is not defined
print(repr(ap)) # Calls __repr__

print("\n--- Example with only __str__ defined ---")
class SimplePoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"SimplePoint at ({self.x}, {self.y})"

sp = SimplePoint(1, 2)
print(sp)       # Calls __str__
print(str(sp))  # Calls __str__
print(repr(sp)) # Calls default object repr because __repr__ is not defined
````

What is the significance of the ‘super()’ function in Python


## What is the significance of the `super()` function in Python?

The `super()` function in Python provides a way to refer to the parent class (or an ancestor class) without explicitly naming it. Its primary significance lies in facilitating cooperative multiple inheritance, ensuring that methods in the Method Resolution Order (MRO) are called correctly.

### Key Significance and Use Cases:

1.  **Calling Parent Class Methods (especially `__init__`):**
    The most common use case for `super()` is to call the `__init__` method of a parent class from a child class's `__init__` method. This ensures that the parent class's attributes are properly initialized.

2.  **Cooperative Multiple Inheritance:**
    In a complex inheritance hierarchy (especially with multiple inheritance), `super()` plays a crucial role. It doesn't just call the *immediate* parent; instead, it calls the next method in the Method Resolution Order (MRO). This mechanism ensures that all parent classes in the inheritance chain have their methods (e.g., constructors or other shared methods) executed, preventing redundancy and ensuring correct initialization or behavior across the hierarchy.

3.  **Maintaining Dynamic Method Dispatch:**
    By using `super()`, your code becomes more flexible and robust to changes in the class hierarchy. If you change a parent class or the MRO, `super()` will automatically adapt, calling the correct next method in the chain, whereas explicitly calling `ParentClass.method(self, ...)` would require manual updates.

4.  **Abstracting Parent Class Name:**
    It helps to make your code cleaner and more maintainable by avoiding hardcoding parent class names. If the parent class changes, you don't need to update every `__init__` or method call in the child classes.

### How `super()` works:

When you call `super()`, it returns a proxy object that delegates method calls to the parent or sibling class of a type. In Python 3, `super()` can be called without arguments inside a method of a class to automatically get the current class and instance/object.

-   **`super().__init__(...)`**: Calls the `__init__` method of the next class in the MRO.
-   **`super().some_method(...)`**: Calls `some_method` of the next class in the MRO.

### Example:

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Person __init__ called for {self.name}")

    def introduce(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old."

class Student(Person):
    def __init__(self, name, age, student_id):
        # Call the parent class's __init__ method
        super().__init__(name, age)
        self.student_id = student_id
        print(f"Student __init__ called for {self.name}")

    def study(self):
        return f"{self.name} (ID: {self.student_id}) is studying."

    def introduce(self):
        # Extend the parent's introduce method
        parent_intro = super().introduce()
        return f"{parent_intro} I am also a student with ID {self.student_id}."

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject
        print(f"Teacher __init__ called for {self.name}")

    def teach(self):
        return f"{self.name} teaches {self.subject}."

# --- Usage Examples ---

student1 = Student("Alice", 20, "S12345")
teacher1 = Teacher("Mr. Smith", 45, "Math")

print(f"\n{student1.introduce()}")
print(f"{student1.study()}")

print(f"\n{teacher1.introduce()}")
print(f"{teacher1.teach()}")

# --- Example with Multiple Inheritance (demonstrating MRO) ---

class Flyer:
    def __init__(self, wingspan):
        self.wingspan = wingspan
        print(f"Flyer __init__ called with wingspan {self.wingspan}")

    def fly(self):
        return f"I can fly with a wingspan of {self.wingspan}m."

class Bird(Flyer):
    def __init__(self, name, wingspan):
        super().__init__(wingspan) # Calls Flyer's __init__
        self.name = name
        print(f"Bird __init__ called for {self.name}")

    def chirpr(self):
        return f"{self.name} chirps!"

class Bat(Flyer):
    def __init__(self, name, wingspan, nocturnal):
        super().__init__(wingspan) # Calls Flyer's __init__
        self.name = name
        self.nocturnal = nocturnal
        print(f"Bat __init__ called for {self.name}")

    def sound(self):
        return f"{self.name} makes a screeching sound!"

# What if we have a FlyingPerson (for example, a superhero)?
class FlyingPerson(Student, Flyer):
    def __init__(self, name, age, student_id, wingspan, superpower):
        # super() ensures that all __init__ methods in the MRO are called
        # Student's __init__ will be called first, then Person's, then Flyer's
        super().__init__(name, age, student_id) # Calls Student's __init__
        # However, Student.__init__ calls Person.__init__ but not Flyer.__init__.
        # In cooperative multiple inheritance, super() in Student's __init__ would be crucial if Person also inherited from something else.
        # For direct multiple inheritance, explicitly calling all parent constructors can be clearer or using 'super()' carefully.

        # A more explicit way for multiple inheritance when __init__ signatures differ:
        Student.__init__(self, name, age, student_id) # Initializes name, age, student_id
        Flyer.__init__(self, wingspan) # Initializes wingspan

        self.superpower = superpower
        print(f"FlyingPerson __init__ called for {self.name}")

    def perform_superpower(self):
        return f"{self.name} uses their {self.superpower}!"

    def introduce(self):
        # Calls Student's introduce, which in turn calls Person's introduce
        student_intro = super().introduce()
        return f"{student_intro} I also {self.fly()} and {self.perform_superpower()}"


print("\n--- MRO for FlyingPerson ---")
print(FlyingPerson.__mro__)

superhero = FlyingPerson("Clark", 25, "S007", 10, "Flight")

# Notice the output of __init__ calls. Without careful super() usage or explicit calls,
# some parent constructors might not be called in multiple inheritance if signatures differ.
# The example above uses explicit calls to demonstrate the initialization.

print(f"\n{superhero.introduce()}")
print(f"{superhero.study()}")
print(f"{superhero.fly()}")
print(f"{superhero.perform_superpower()}")
````

In the `Student` and `Teacher` classes, `super().__init__(name, age)` ensures that the `Person` class's initialization logic is run, setting up `name` and `age`. Similarly, `super().introduce()` allows the `Student` class to extend the `Person`'s `introduce` method rather than completely rewriting it.

For `FlyingPerson`, demonstrating cooperative multiple inheritance with `super()` can be tricky with differing `__init__` signatures. Often, explicit calls to parent constructors (`ParentClass.__init__(self, ...)`) are used for clarity in such cases, or `super()` is used with a carefully designed `__init__` method that accepts `*args` and `**kwargs` and passes them up the chain. The `FlyingPerson` example explicitly calls the parent `__init__` methods for clarity due to differing argument requirements.

What is the significance of the __del__ method in Python


## What is the significance of the `__del__` method in Python?

The `__del__` method, also known as the **destructor** or **finalizer**, is a special method in Python classes. It is called when an instance of a class is about to be *destroyed* (i.e., its reference count drops to zero and it's garbage collected). Its primary significance is in performing cleanup operations before an object is completely removed from memory.

### Key Significance and Use Cases:

1.  **Resource Cleanup:** The most common use case for `__del__` is to release external resources that the object might be holding. This includes:
    *   Closing files (though `with` statements are generally preferred for files).
    *   Releasing network connections.
    *   Freeing up database connections.
    *   Releasing locks.
    *   Deleting temporary files.

2.  **Notification/Logging:** It can be used to log or notify when an object is being destroyed, which can be useful for debugging or monitoring resource usage.

### Important Considerations and Limitations:

Despite its apparent purpose, `__del__` is **rarely used** in modern Python programming, and often actively discouraged for several reasons:

-   **Unpredictable Execution:** Python's garbage collector determines when an object is no longer referenced and can be collected. The timing of `__del__` execution is therefore **not guaranteed** and can be unpredictable. An object might persist in memory longer than expected, or `__del__` might not be called at all if the program terminates abruptly or if circular references prevent garbage collection.
-   **Order of Destruction:** In complex applications, the order in which objects are destroyed (and thus `__del__` methods are called) is not guaranteed, which can lead to situations where an object's `__del__` tries to use another object that has already been destroyed.
-   **Exceptions in `__del__`:** If an exception occurs within a `__del__` method, it is typically ignored by Python (or printed as a warning to `sys.stderr`), and it can lead to silent failures or unexpected behavior.
-   **Better Alternatives (`with` statement):** For managing resources like files, network connections, or locks, the `with` statement (context managers) is the preferred and more Pythonic approach. Context managers ensure resources are properly acquired and released, even if errors occur.
-   **Circular References:** Objects involved in circular references might not be garbage collected immediately (or ever, in older Python versions without a cycle detector), preventing `__del__` from being called.

**In summary, while `__del__` provides a mechanism for finalization, its non-deterministic nature makes it unsuitable for critical resource management. Context managers are almost always the better choice for ensuring resources are properly released.**

### Example:

````python
import time

class ResourceUser:
    def __init__(self, name):
        self.name = name
        print(f"ResourceUser '{self.name}' created.")
        # Simulate acquiring a resource
        self.resource = open(f"temp_{self.name}.txt", "w")
        self.resource.write(f"Data for {self.name}\n")
        print(f"File 'temp_{self.name}.txt' opened.")

    def do_work(self):
        print(f"ResourceUser '{self.name}' is doing some work.")

    def __del__(self):
        # This method is called when the object is garbage collected
        if hasattr(self, 'resource') and not self.resource.closed:
            self.resource.close()
            print(f"ResourceUser '{self.name}' __del__ called: File 'temp_{self.name}.txt' closed.")
        else:
            print(f"ResourceUser '{self.name}' __del__ called: No resource to close or already closed.")

# Create an instance
user1 = ResourceUser("Logger")
user1.do_work()

# Explicitly delete the reference. This makes the object eligible for garbage collection.
# The __del__ method *might* be called immediately, but it's not guaranteed.
del user1

print("\n--- End of script (garbage collection might happen now or later) ---")
# The garbage collector might run at the end of the script or at other times.
# You might see the __del__ message here, or it might have already printed.

# Another example without explicit del, letting GC handle it naturally
print("\nCreating another user...")
user2 = ResourceUser("NetworkClient")
user2.do_work()

# When user2 goes out of scope (e.g., end of function or program),
# or when garbage collection runs, __del__ will be called.
# We'll pause briefly to let any pending GC happen before the program ends.
time.sleep(0.1)
````

In the example:
- When `user1` is created, `__init__` is called.
- When `del user1` is executed (removing the last reference to the object), the `ResourceUser` object becomes eligible for garbage collection, and its `__del__` method is likely called soon after.
- For `user2`, `__del__` is called when the program finishes and the object is cleaned up by the garbage collector (or earlier if the interpreter decides).

## Class Methods

A **class method** is a method that is bound to the class and not the instance of the class. It receives the class itself as its first argument (conventionally named `cls`), instead of the instance (`self`). Class methods are defined using the `@classmethod` decorator.

### When to use class methods:
-   **Factory Methods:** To create instances of the class in different ways.
-   **Alternative Constructors:** When you need multiple ways to initialize an object.
-   **Modifying Class State:** To perform operations that modify a class variable that applies to all instances.

````python
class Car:
    wheels = 4  # Class variable

    def __init__(self, make, model):
        self.make = make
        self.model = model

    def display_info(self):
        return f"This is a {self.make} {self.model} with {self.wheels} wheels."

    @classmethod
    def with_custom_wheels(cls, make, model, num_wheels):
        car = cls(make, model) # Create an instance of the class
        car.wheels = num_wheels # Modify instance-specific 'wheels' (or class if not instance)
        return car

    @classmethod
    def change_default_wheels(cls, new_num_wheels):
        cls.wheels = new_num_wheels # Modifies the class variable 'wheels'
        return f"Default car now has {cls.wheels} wheels."

# Using instance method
car1 = Car("Toyota", "Camry")
print(car1.display_info())

# Using class method as a factory
car2 = Car.with_custom_wheels("Honda", "Civic", 3)
print(car2.display_info()) # This will show 3 wheels for car2

# What about car1? It still has 4 wheels as 'wheels' was set on the instance
print(car1.display_info())

# Change the class-level default wheels
print(Car.change_default_wheels(6))

# Now, new cars will have 6 wheels by default
car3 = Car("Ford", "Focus")
print(car3.display_info())

# car1 and car2 are unaffected by the class-level change unless they didn't have their own 'wheels' attribute
# Since with_custom_wheels explicitly sets car.wheels, car2 is still 3
# car1 also has its own 'wheels' attribute set during its initialization or earlier, if not, it would inherit
````

## Static Methods

A **static method** is a method that belongs to a class but doesn't operate on the instance or the class itself. It doesn't receive `self` or `cls` as its first argument. Static methods are defined using the `@staticmethod` decorator.

### When to use static methods:
-   **Utility Functions:** When a method has a logical connection to the class but doesn't need to access any instance or class-specific data.
-   **Helper Functions:** For functions that perform independent tasks but are conceptually related to the class.

````python
import datetime

class MathOperations:
    def __init__(self, value):
        self.value = value

    @staticmethod
    def add_numbers(x, y):
        return x + y

    @staticmethod
    def is_weekend(date):
        return date.weekday() >= 5  # Saturday (5) or Sunday (6)

    def multiply_by_self(self):
        return self.value * self.value

# Using static methods directly through the class or an instance
print(f"Sum: {MathOperations.add_numbers(5, 7)}")

today = datetime.date(2023, 10, 28) # A Saturday
print(f"Is {today} a weekend? {MathOperations.is_weekend(today)}")

# Using an instance method
calc = MathOperations(10)
print(f"Multiply by self: {calc.multiply_by_self()}")
````

## Key Differences:

| Feature             | Instance Methods                 | Class Methods                    | Static Methods                   |
|---------------------|----------------------------------|----------------------------------|----------------------------------|
| **First Argument**  | `self` (instance)                | `cls` (class)                    | None (`self` or `cls`)           |
| **Decorator**       | None (default)                   | `@classmethod`                   | `@staticmethod`                  |
| **Access to State** | Can access/modify instance state | Can access/modify class state    | Cannot access instance or class state |
| **Use Case**        | Object-specific operations       | Factory methods, class-level logic | Utility functions, no state dependency |
| **Calling**         | `obj.method()`                   | `Class.method()` or `obj.method()` | `Class.method()` or `obj.method()` |

What is method chaining in Python OOP


## What is Method Chaining in Python OOP?

**Method chaining** is a technique in Object-Oriented Programming (OOP) where multiple method calls are invoked sequentially on the same object. Each method in the chain performs an operation and returns a reference to the object itself (or a modified version of it), allowing the next method call to be made directly on the result of the previous one. This creates a concise and readable sequence of operations.

The core idea is that methods are designed to be *fluent*, meaning they return `self` (the instance itself) after performing their action, rather than `None` or some other value, which would break the chain.

### Benefits of Method Chaining:

1.  **Readability and Conciseness:** It allows you to express a sequence of operations on an object in a single line or a clear, vertical flow, which can make the code easier to read and understand than nested function calls or separate statements.
2.  **Reduced Temporary Variables:** It eliminates the need for creating intermediate variables to store the result of each step in a sequence of operations.
3.  **Fluency:** It creates a more fluent and expressive API for classes, mimicking natural language when describing a series of actions on an object.
4.  **Immutability (Optional but common):** Often, in more advanced use cases (like with pandas DataFrames or certain builder patterns), each method returns a *new* modified object instead of modifying the original in-place. This promotes immutability, which can reduce side effects and make debugging easier.

````python
class Calculator:
    def __init__(self, value=0):
        self.value = value

    def add(self, num):
        self.value += num
        return self  # Return self to allow chaining

    def subtract(self, num):
        self.value -= num
        return self  # Return self to allow chaining

    def multiply(self, num):
        self.value *= num
        return self  # Return self to allow chaining

    def divide(self, num):
        if num == 0:
            raise ValueError("Cannot divide by zero")
        self.value /= num
        return self  # Return self to allow chaining

    def get_result(self):
        return self.value

# --- Demonstrating Method Chaining ---

# Start with a value of 10, add 5, multiply by 2, subtract 3
calc = Calculator(10)
result = calc.add(5).multiply(2).subtract(3).get_result()
print(f"Chained calculation result: {result}")

# Another chain: start with 0, add 10, divide by 5, add 7
calc2 = Calculator()
result2 = calc2.add(10).divide(5).add(7).get_result()
print(f"Another chained calculation result: {result2}")

# Without chaining, it would look like this:
calc3 = Calculator(10)
calc3.add(5)
calc3.multiply(2)
calc3.subtract(3)
result3 = calc3.get_result()
print(f"Non-chained calculation result: {result3}")
````

In this example:
- Each arithmetic method (`add`, `subtract`, `multiply`, `divide`) modifies the `value` attribute of the `Calculator` object and then returns `self`.
- Because `self` (the `Calculator` instance) is returned, we can immediately call another method on that same instance, thus chaining the operations: `calc.add(5).multiply(2).subtract(3)`.
- The `get_result()` method is then called at the end of the chain to retrieve the final computed value.

## What is the purpose of the `__call__` method in Python?

The `__call__` method is a special method (a dunder method) in Python classes that allows an instance of a class to be called as a function. If a class implements the `__call__` method, then its objects can be used like regular functions, allowing you to pass arguments to them and receive a return value.

### Key Purpose and Use Cases:

1.  **Callable Objects (Functors):** It makes instances of a class "callable." This is useful when you want an object to encapsulate some state and behavior, but also want to use it with function-like syntax.
2.  **Stateful Decorators:** Decorators in Python are functions that take another function as an argument and return a new function. If you need a decorator to maintain state across multiple calls to the decorated function, you can implement it as a class with a `__call__` method.
3.  **Custom Callbacks:** When designing APIs or libraries where users might want to define custom logic that gets invoked with specific arguments, providing a callable object can be a flexible alternative to passing a simple function.
4.  **Implementing Strategies or Policies:** An object with `__call__` can represent a specific strategy or policy that can be invoked at different points in a program, carrying its own configuration or state.
5.  **Simulating Closures with More Control:** While closures are often sufficient, `__call__` provides a more structured way to create function-like objects that carry state and can have other methods.

### How it works:

When you call an object `obj` like a function (e.g., `obj(arg1, arg2)`), Python automatically looks for and executes the `__call__` method of that object's class. The arguments passed to the object call are then passed directly to the `__call__` method.

### Example:

````python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
        print(f"Multiplier instance created with factor: {self.factor}")

    def __call__(self, number):
        print(f"__call__ method invoked with number: {number}")
        return number * self.factor

# Create an instance of the Multiplier class
double = Multiplier(2)
triple = Multiplier(3)

# Now, call the instances as if they were functions
result1 = double(5) # This implicitly calls double.__call__(5)
print(f"Double of 5: {result1}")

result2 = triple(10) # This implicitly calls triple.__call__(10)
print(f"Triple of 10: {result2}")

result3 = double(7)
print(f"Double of 7: {result3}")

# Example as a stateful decorator
class CountCalls:
    def __init__(self, func):
        self.func = func
        self.num_calls = 0

    def __call__(self, *args, **kwargs):
        self.num_calls += 1
        print(f"Calling {self.func.__name__}, call count: {self.num_calls}")
        return self.func(*args, **kwargs)

@CountCalls
def say_hello(name):
    print(f"Hello, {name}!")

@CountCalls
def add(a, b):
    return a + b

say_hello("Alice")
say_hello("Bob")

sum_result = add(10, 20)
print(f"Sum: {sum_result}")
sum_result = add(5, 5)
print(f"Sum: {sum_result}")

print(f"say_hello was called {say_hello.num_calls} times")
print(f"add was called {add.num_calls} times")
````

In the `Multiplier` example:
- We create instances `double` and `triple` by passing a `factor` during initialization.
- When we call `double(5)`, the `__call__` method of the `double` object is executed, multiplying `5` by the stored `factor` (`2`).

In the `CountCalls` decorator example:
- The `CountCalls` class acts as a decorator. When `@CountCalls` is applied to `say_hello` and `add`, it effectively replaces the original function with an instance of `CountCalls` that wraps the function.
- When `say_hello("Alice")` is called, it's actually `CountCalls_instance_for_say_hello.__call__("Alice")` that gets executed, incrementing the call counter and then calling the original `say_hello` function.
- This demonstrates how `__call__` allows an object to maintain state (`num_calls`) and execute function-like behavior.

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        """Developer-friendly representation."""
        return f"Vector({self.x}, {self.y})"

    def __str__(self):
        """User-friendly representation."""
        return f"<{self.x}, {self.y}>"

    def __add__(self, other):
        """Overrides the + operator for vector addition."""
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Can only add a Vector object to another Vector object")

# --- Test the Vector class ---

v1 = Vector(2, 3)
v2 = Vector(4, 1)
v3 = Vector(-1, 5)

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Vector 3: {v3}")

# Perform vector addition using the + operator
v_sum1 = v1 + v2
print(f"Vector 1 + Vector 2 = {v_sum1}") # Expected: <6, 4>

v_sum2 = v1 + v3
print(f"Vector 1 + Vector 3 = {v_sum2}") # Expected: <1, 8>

# Adding multiple vectors
v_total = v1 + v2 + v3
print(f"Vector 1 + Vector 2 + Vector 3 = {v_total}") # Expected: <5, 9>

# Attempt to add non-Vector object (will raise TypeError)
try:
    v1 + 5
except TypeError as e:
    print(f"\nError: {e}")
````

````python
class Book:
    def __init__(self, title, author, year_published):
        if not isinstance(title, str) or not title.strip():
            raise ValueError("Title cannot be empty.")
        if not isinstance(author, str) or not author.strip():
            raise ValueError("Author cannot be empty.")
        if not isinstance(year_published, int) or year_published <= 0:
            raise ValueError("Year published must be a positive integer.")

        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        """Returns a formatted string with the book's details."""
        return f"\" {self.title}\" by {self.author}, published in {self.year_published}."

# --- Test the Book class ---

book1 = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 1979)
book2 = Book("1984", "George Orwell", 1949)
book3 = Book("Pride and Prejudice", "Jane Austen", 1813)

print(book1.get_book_info())
print(book2.get_book_info())
print(book3.get_book_info())

# Example of invalid input
try:
    Book("", "Unknown Author", 2000) # Empty title
except ValueError as e:
    print(f"\nError creating book: {e}")

try:
    Book("Valid Title", "", 2000) # Empty author
except ValueError as e:
    print(f"Error creating book: {e}")

try:
    Book("Valid Title", "Valid Author", -10) # Invalid year
except ValueError as e:
    print(f"Error creating book: {e}")
````

18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.  


````python
class House:
    def __init__(self, address, price):
        if not isinstance(address, str) or not address.strip():
            raise ValueError("Address cannot be empty.")
        if not isinstance(price, (int, float)) or price <= 0:
            raise ValueError("Price must be a positive number.")

        self.address = address
        self.price = price

    def get_house_info(self):
        return f"Address: {self.address}, Price: ${self.price:,.2f}"

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price) # Call parent class constructor
        if not isinstance(number_of_rooms, int) or number_of_rooms <= 0:
            raise ValueError("Number of rooms must be a positive integer.")

        self.number_of_rooms = number_of_rooms

    def get_mansion_info(self):
        base_info = super().get_house_info()
        return f"{base_info}, Rooms: {self.number_of_rooms}"

# --- Test the classes ---

house1 = House("123 Main St", 350000)
mansion1 = Mansion("456 Oak Ave", 1200000, 10)

print(f"House Info: {house1.get_house_info()}")
print(f"Mansion Info: {mansion1.get_mansion_info()}")

print(f"\nMansion address: {mansion1.address}")
print(f"Mansion price: ${mansion1.price:,.2f}")
print(f"Mansion rooms: {mansion1.number_of_rooms}")

# Example of invalid input
try:
    House("", 200000) # Empty address
except ValueError as e:
    print(f"\nError creating house: {e}")

try:
    Mansion("789 Pine Ln", 500000, -5) # Invalid number of rooms
except ValueError as e:
    print(f"Error creating mansion: {e}")
````
